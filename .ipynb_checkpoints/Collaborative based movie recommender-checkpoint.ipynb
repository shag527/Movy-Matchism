```python
# Importing the required libraries

import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Merging the 2 datasets - ratings and names of the movies

ratings=pd.read_csv('/home/shagun/Documents/Ml Projects/MRS/Movy-Matchism/Datasets/movielens_dataset/ratings.csv')
movies=pd.read_csv('/home/shagun/Documents/Ml Projects/MRS/Movy-Matchism/Datasets/movielens_dataset/movies.csv')
data=pd.merge(ratings,movies).drop(['timestamp'],axis=1)
data.head()
```

```python
# creating a pivot table

user_ratings=data.pivot_table(index=['userId'],columns=['title'],values='rating')
user_ratings.head()
```

```python
#Removing the movies which have less than 10 users who rated them and fill the remaining with zero

user_ratings=user_ratings.dropna(thresh=10,axis=1)
user_ratings=user_ratings.fillna(value=0,axis=1)
user_ratings.head()
```

```python
#building the similarity matrix using pearson's correlation

item_similarity=user_ratings.corr(method='pearson')
item_similarity.head()
```

```python
#Getting score of similar movies

def get_similar_movies(movie_name,user_rating):
    score=item_similarity[movie_name]*(user_rating-2.5)
    score=score.sort_values(ascending=False)
    
    return score
```

```python
user_list=[("(500) Days of Summer (2009)",5),("10,000 BC (2008)",3)]

similar_movies=pd.DataFrame()

for movie,rating in user_list:
    similar_movies=similar_movies.append(get_similar_movies(movie,rating),ignore_index=True)
    
similar_movies.head()
similar_movies.sum().sort_values(ascending=False)
```
